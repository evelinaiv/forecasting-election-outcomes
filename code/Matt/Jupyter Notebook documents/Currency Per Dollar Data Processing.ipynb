```python
import sqlite3 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
CurrencyPerDollar = pd.read_csv(r'C:\Users\Matth\Downloads\USD Exchange Rates.csv')
```

```python
#Setting Country name as the index
CurrencyPerDollar.set_index(CurrencyPerDollar.columns[0], inplace=True)
```

```python
#Removing junk rows
CurrencyPerDollar.drop(CurrencyPerDollar.columns[0:3], axis=1, inplace=True)
```

```python
#Moving the years to index
CurrencyPerDollar = CurrencyPerDollar.transpose()
```

```python
CurrencyPerDollar.head()
```

```python
CurrencyPerDollar.info()
```

```python
#Removing countries with little to no data

non_null_count = CurrencyPerDollar.count()

non_null_perc = (non_null_count/len(CurrencyPerDollar)) *100

columns_to_drop = non_null_perc[non_null_perc <= 10].index

CurrencyPerDollar = CurrencyPerDollar.drop(columns=columns_to_drop)

CurrencyPerDollar.head()
```

```python
#Now we a nice and usable dataframe
```

```python
CurrencyPerDollar.to_csv('Currency Per Dollar.csv', index=True)
```
