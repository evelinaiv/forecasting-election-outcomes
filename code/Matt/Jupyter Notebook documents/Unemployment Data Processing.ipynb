```python
import sqlite3 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
Unemployment = pd.read_csv(r'C:\Users\Matth\Downloads\Unemployment.csv')
```

```python
#Setting Country name as the index
Unemployment.set_index(Unemployment.columns[0], inplace=True)
```

```python
#Removing junk rows
Unemployment.drop(Unemployment.columns[0:3], axis=1, inplace=True)
```

```python
#Moving the years to index
Unemployment = Unemployment.transpose()
```

```python
#For Unemployment there is no data until 1990 so this is removed
Unemployment.drop(Unemployment.index[0:31], axis=0, inplace=True)
```

```python
Unemployment.head()
```

```python
#Removing countries with little to no data

non_null_count = Unemployment.count()

non_null_perc = (non_null_count/len(Unemployment)) *100

columns_to_drop = non_null_perc[non_null_perc <= 10].index

Unemployment = Unemployment.drop(columns=columns_to_drop)

Unemployment.head()
```

```python
#Creating a dataset which contains percentage changes in inflation in each year
Unemployment_perc_change = Unemployment.pct_change() * 100
Unemployment_perc_change.head()
```

```python
#Now we have two nice and usable dataframes
```

```python
Unemployment.to_csv('Unemployment.csv', index=True)
Unemployment_perc_change.to_csv('Unemployment_perc_change.csv', index=True)
```
