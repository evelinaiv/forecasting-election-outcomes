```python
import sqlite3 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
VAT = pd.read_excel(r"C:\Users\Matth\Downloads\VAT.xlsx")
```

```python
#Setting Country name as the index
VAT.set_index(VAT.columns[0], inplace=True)
VAT.head()
```

```python
#Removing Junk rows
VAT.drop(VAT.columns[20:24], axis=1, inplace=True)
VAT.drop(VAT.columns[18], axis=1, inplace=True)
VAT.drop(VAT.columns[16], axis=1, inplace=True)
```

```python
VAT = VAT.transpose()
VAT.head()
```

```python
#Changing "-" into NaN values

VAT.replace({' - ': np.nan}, inplace=True)
```

```python
#Removing countries with little to no data

non_null_count = VAT.count()

non_null_perc = (non_null_count/len(VAT)) *100

columns_to_drop = non_null_perc[non_null_perc <= 10].index

VAT = VAT.drop(columns=columns_to_drop)

VAT.head()
```

```python
#Standardising column names by removing "*" for countries to match column names with other datasets
VAT.columns = VAT.columns.str.replace(r'\*$', '', regex=True)
VAT.head()
```

```python
#Creating a dataset which contains percentage changes in inflation in each year
VAT_perc_change = VAT.pct_change() * 100
VAT_perc_change.head()
```

```python
#Now we have 2 Dataframes with VAT data, one with the levels of it and one with any percentage cahnges in each year
```

```python
VAT.to_csv('VAT.csv', index=True)
VAT_perc_change.to_csv('VAT_perc_change.csv', index=True)
```
