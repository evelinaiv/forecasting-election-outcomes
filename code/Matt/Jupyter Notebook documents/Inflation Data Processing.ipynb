```python
import sqlite3 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
Inflation = pd.read_csv(r'C:\Users\Matth\Downloads\Inflation.csv')
```

```python
#Setting Country name as the index
Inflation.set_index(Inflation.columns[0], inplace=True)
```

```python
#Removing junk rows
Inflation.drop(Inflation.columns[0:3], axis=1, inplace=True)
```

```python
#Moving the years to index
Inflation = Inflation.transpose()
```

```python
Inflation.head()
```

```python
Inflation.info()
```

```python
#Removing countries with little to no data

non_null_count = Inflation.count()

non_null_perc = (non_null_count/len(Inflation)) *100

columns_to_drop = non_null_perc[non_null_perc <= 10].index

Inflation = Inflation.drop(columns=columns_to_drop)

Inflation.head()
```

```python
#Creating a dataset which contains percentage changes in inflation in each year
Inflation_perc_change = Inflation.pct_change() * 100
Inflation_perc_change.head()
```

```python
#Now we have two nice and usable dataframes
```

```python
Inflation_perc_change.to_csv('Inflation_perc_change.csv', index=True)
Inflation.to_csv('Inflation.csv', index=True)
```
